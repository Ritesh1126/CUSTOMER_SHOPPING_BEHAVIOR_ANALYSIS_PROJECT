```python
### checking if the null value is getting there , If yes then group & transform by using lambda function , mean, median .
# for example , we checking for the --> Review Rating for analysis
df['Review Rating'] = df.groupby('Category')['Review Rating'].transform(lambda x: x.fillna(x.median()))
```

```python
df.isnull().sum()
```

```python
df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(' ', '_')
df = df.rename(columns={'purchase_amount_(usd)': 'purchase_amount'})
df.columns
##  we replace , Capital letter cases  with smmaller/lower cases and "  " with the " _ ".
##  Cause of easy to take a reference for 'sql' implementing...
```

```python
# create a column of age_group
labels = ['Young Adult', 'Middle-age' , 'Adult', 'Senior']
df['age_group'] = pd.qcut(df['age'], q=4, labels = labels)
df[['age', 'age_group']].head(10)
```

```python
 #create column purchase_frequency_days
frequency_mapping = {
     'fortnightly' : 14,
     'weekly' : 7,
     'monthly' : 30,
     'every 3 months' : 90,
     'quaterly' : 90,
     'yearly' : 365,
     'never' : 0,
     'unknown' : 0
 }
df['purchase_frequency_days'] = df['frequency_of_purchases'].map(frequency_mapping)
df[['frequency_of_purchases', 'purchase_frequency_days']].head(10)
```
